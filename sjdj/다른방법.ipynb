```python
from selenium import webdriver  # type: ignore
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pyperclip
import time
import pandas as pd
import os
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, ElementClickInterceptedException, StaleElementReferenceException

# âœ… ë¸Œë¼ìš°ì € ì°½ í¬ê¸° ì„¤ì •
WINDOW_WIDTH = 1100
WINDOW_HEIGHT = 1000

# âœ… ë¸Œë¼ìš°ì € ì´ˆê¸°í™” ë° ë¡œê·¸ì¸ (ì¤Œ ë ˆë²¨ 25% ì„¤ì • ì¶”ê°€)
def setup_browser(position_x):
    driver = webdriver.Chrome()
    driver.set_window_size(WINDOW_WIDTH, WINDOW_HEIGHT)
    driver.set_window_position(position_x, 0)
    driver.get('https://console.thebackend.io/ko/login')

    time.sleep(0.2)
    return driver

# âœ… ë¡œê·¸ì¸ í•¨ìˆ˜
def login_to_console(driver):
    try:
        id_input = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "username"))
        )
        pyperclip.copy("dksrufp0607@naver.com")
        id_input.send_keys(Keys.CONTROL, 'v')

        pw_input = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "password"))
        )
        pyperclip.copy("ghkfkd159!")
        pw_input.send_keys(Keys.CONTROL, 'v')
        time.sleep(1)
        login_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[type="submit"]'))
        )
        login_button.click()
        print("âœ… ë¡œê·¸ì¸ ì™„ë£Œ!")
        time.sleep(2)
    except Exception as e:
        print(f"âš ï¸ ë¡œê·¸ì¸ ì‹¤íŒ¨: {e}")

# âœ… ì½˜ì†” í˜ì´ì§€ ì´ë™
def navigate_to_page(driver):
    article_url = 'https://console.thebackend.io/ko/project/1ea3f14d34e89530ea88b3245bc82dc17d5f52ce1554049f19fce9219a847cfce18bb8891c9ffe90bc65e2b9a3b981853fc5513c1dd200afc9590ba6bfd5fced4230647d25328849e0917641/baseGameInfo/data'
    driver.get(article_url)
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.TAG_NAME, "body"))
    )
    print("âœ… í˜ì´ì§€ ë¡œë“œ ì™„ë£Œ!")

# âœ… "í™•ì¸" ë²„íŠ¼ í´ë¦­
def confirm_click(driver):
    try:
        print("ğŸ” 'í™•ì¸' ë²„íŠ¼ íƒìƒ‰ ì¤‘...")
        confirm_button = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "í™•ì¸")]'))
        )
        confirm_button.click()
        print("âœ… í™•ì¸ ë²„íŠ¼ í´ë¦­ ì™„ë£Œ!")
    except TimeoutException:
        print("âš ï¸ 'í™•ì¸' ë²„íŠ¼ì„ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. í˜ì´ì§€ê°€ ì œëŒ€ë¡œ ë¡œë“œë˜ì—ˆëŠ”ì§€ í™•ì¸í•˜ì„¸ìš”.")
    except Exception as e:
        print(f"âš ï¸ í™•ì¸ ë²„íŠ¼ í´ë¦­ ì‹¤íŒ¨: {e}")

# âœ… CSV ë°ì´í„° ë¡œë“œ í•¨ìˆ˜ (ê²€ìƒ‰ìš©)
def get_csv_value(row, col):
    csv_path = "C:/ì—‘ì…€/Ranking_1.csv"
    try:
        df = pd.read_csv(csv_path)
        if row < df.shape[0] and col < df.shape[1]:
            print(f"ğŸ”¹ {row}í–‰ 3ì—´ ê°’ (ì¶œë ¥ìš©): {df.iloc[row, 2]}")
            return df.iloc[row, col]
        else:
            print("âš ï¸ CSV íŒŒì¼ì˜ í¬ê¸°ë¥¼ ì´ˆê³¼í•˜ëŠ” ì¸ë±ìŠ¤ì…ë‹ˆë‹¤.")
            return None
    except FileNotFoundError:
        print(f"âš ï¸ CSV íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {csv_path}")
        return None
    except Exception as e:
        print(f"âš ï¸ CSV ë¡œë“œ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return None

# âœ… ê²€ìƒ‰ ì‹¤í–‰ í•¨ìˆ˜
def execute_tasks(driver, row, col):
    try:
        b2_value = get_csv_value(row, col)
        if not b2_value:
            print("âš ï¸ CSVì—ì„œ ê°’ì„ ê°€ì ¸ì˜¤ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. ì‘ì—…ì„ ì¤‘ë‹¨í•©ë‹ˆë‹¤.")
            return

        table_dropdown = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//div[contains(@class, "_manageInfo-data_1rntr_71")]'))
        )
        table_dropdown.click()
        print("âœ… í…Œì´ë¸” ë“œë¡­ë‹¤ìš´ í´ë¦­ ì™„ë£Œ!")

        base_data_option = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//div[@role="option"][span[text()="BASE_DATA"]]'))
        )
        base_data_option.click()
        print("âœ… BASE_DATA ì„ íƒ ì™„ë£Œ!")

        detail_search_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//p[contains(text(), "ìƒì„¸ ê²€ìƒ‰")]'))
        )
        time.sleep(3)
        try:
            detail_search_button.click()
        except ElementClickInterceptedException:
            print("âš ï¸ 'ìƒì„¸ ê²€ìƒ‰' ë²„íŠ¼ í´ë¦­ ì°¨ë‹¨ë¨. JavaScriptë¡œ ì¬ì‹œë„...")
            driver.execute_script("arguments[0].click();", detail_search_button)
        print("âœ… ìƒì„¸ ê²€ìƒ‰ í´ë¦­ ì™„ë£Œ!")

        search_input = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.NAME, "defaultSearchValue"))
        )
        search_input.clear()
        search_input.send_keys(b2_value)
        print(f"âœ… B2 ê°’ ì…ë ¥ ì™„ë£Œ: {b2_value}")

        search_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(@class, "ui medium positive button") and @type="submit"]'))
        )
        search_button.click()
        print("âœ… ê²€ìƒ‰ ë²„íŠ¼ í´ë¦­ ì™„ë£Œ!")

        uuid_element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//p[contains(@class, "_link_a1mor_113")]'))
        )
        uuid_element.click()
        print("âœ… UUID í´ë¦­ ì™„ë£Œ!")
    except Exception as e:
        print(f"âš ï¸ ê²€ìƒ‰ ì‘ì—… ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

# âœ… ì—‘ì…€ì—ì„œ ë°ì´í„° ë¡œë“œ í•¨ìˆ˜
def load_excel_data(filename):
    try:
        df = pd.read_excel(filename)
        data = df[["Column Name", "Data Type", "Value"]].values.tolist()
        print(f"âœ… ì—‘ì…€ ë°ì´í„° ë¡œë“œ ì™„ë£Œ! ê²½ë¡œ: {os.path.abspath(filename)}")
        return data
    except FileNotFoundError:
        print(f"âš ï¸ ì—‘ì…€ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {filename}")
        return []
    except Exception as e:
        print(f"âš ï¸ ì—‘ì…€ íŒŒì¼ ë¡œë“œ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return []

# âœ… Ace Editor ê°’ ì„¤ì • í•¨ìˆ˜ (JavaScript ì‚¬ìš©)
def set_ace_editor_value(driver, ace_element, new_value):
    try:
        WebDriverWait(driver, 10).until(
            lambda d: d.execute_script("return typeof ace !== 'undefined' && ace.edit(arguments[0]) !== null", ace_element)
        )
        current_value = driver.execute_script("""
            var editor = ace.edit(arguments[0]);
            return editor.getValue();
        """, ace_element).strip()
        print(f"ğŸ” í˜„ì¬ ê°’: {current_value}")

        driver.execute_script("""
            var editor = ace.edit(arguments[0]);
            editor.setValue(arguments[1], -1);
            editor.session.getUndoManager().reset();
        """, ace_element, str(new_value))

        updated_value = driver.execute_script("""
            var editor = ace.edit(arguments[0]);
            return editor.getValue();
        """, ace_element).strip()
        print(f"âœ… Ace Editor ê°’ ì„¤ì • ì™„ë£Œ: {updated_value}")
    except Exception as e:
        print(f"âš ï¸ Ace Editor ê°’ ì„¤ì • ì‹¤íŒ¨: {e}")

# âœ… í´ë¦½ë³´ë“œë¡œ ê°’ ì„¤ì • í•¨ìˆ˜ (textareaìš©)
def set_value_with_clipboard(driver, element, new_value):
    try:
        pyperclip.copy(str(new_value))
        driver.execute_script("arguments[0].click();", element)  # ìš”ì†Œ í´ë¦­ìœ¼ë¡œ í¬ì»¤ìŠ¤ ì´ë™
        time.sleep(0.5)  # í¬ì»¤ìŠ¤ í›„ ëŒ€ê¸°
        element.send_keys(Keys.CONTROL, 'a')  # ì „ì²´ ì„ íƒ
        time.sleep(0.5)  # ì„ íƒ í›„ ëŒ€ê¸°
        element.send_keys(Keys.DELETE)  # ê°’ ì§€ìš°ê¸°
        time.sleep(0.5)  # ì§€ìš°ê¸° í›„ ëŒ€ê¸°
        element.send_keys(Keys.CONTROL, 'v')  # í´ë¦½ë³´ë“œì—ì„œ ë¶™ì—¬ë„£ê¸°
        print(f"âœ… í´ë¦½ë³´ë“œë¡œ ê°’ ì„¤ì • ì™„ë£Œ: {new_value}")
    except Exception as e:
        print(f"âš ï¸ í´ë¦½ë³´ë“œ ê°’ ì„¤ì • ì‹¤íŒ¨: {e}")

# âœ… ë°ì´í„° ë¶™ì—¬ë„£ê¸° í•¨ìˆ˜ (í•˜ì´ë¸Œë¦¬ë“œ ë°©ì‹ + ìŠ¤í¬ë¡¤ ë° 3ì´ˆ ë”œë ˆì´)
def paste_data(driver, data):
    try:
        # ìˆ˜ì • ë²„íŠ¼ í´ë¦­
        edit_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(@class, "ui mini primary button") and text()="ìˆ˜ì •"]'))
        )
        edit_button.click()
        print("âœ… ìˆ˜ì • ë²„íŠ¼ í´ë¦­ ì™„ë£Œ!")
        time.sleep(1)  # ëª¨ë‹¬ ë¡œë”© ëŒ€ê¸°
            # ì›¹ í™”ë©´ì„ 25%ë¡œ ì¶•ì†Œ
        driver.execute_script("document.body.style.zoom='25%'")
        print("âœ… ì›¹ í™”ë©´ì„ 25%ë¡œ ì¶•ì†Œ ì™„ë£Œ!")

        # í˜ì´ì§€ í•˜ë‹¨ìœ¼ë¡œ ìŠ¤í¬ë¡¤
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        print("âœ… í˜ì´ì§€ í•˜ë‹¨ìœ¼ë¡œ ìŠ¤í¬ë¡¤ ì™„ë£Œ!")
        time.sleep(3)  # ìŠ¤í¬ë¡¤ í›„ 3ì´ˆ ëŒ€ê¸°

        # iframe ë˜ëŠ” ê¸°ë³¸ ì»¨í…ìŠ¤íŠ¸ í™•ì¸
        driver.switch_to.default_content()
        context_found = False

        # ê¸°ë³¸ ì»¨í…ìŠ¤íŠ¸ í™•ì¸
        try:
            WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CLASS_NAME, "_columnBorder-edit-modal_1rntr_47"))
            )
            print("ğŸ” ê¸°ë³¸ ì½˜í…ì¸ ì—ì„œ ìš”ì†Œ ë°œê²¬!")
            context_found = True
        except TimeoutException:
            print("âš ï¸ ê¸°ë³¸ ì½˜í…ì¸ ì—ì„œ ìš”ì†Œ ì—†ìŒ")

        # iframe í™•ì¸
        if not context_found:
            iframes = driver.find_elements(By.TAG_NAME, "iframe")
            for i, frame in enumerate(iframes):
                driver.switch_to.frame(frame)
                try:
                    WebDriverWait(driver, 5).until(
                        EC.presence_of_element_located((By.CLASS_NAME, "_columnBorder-edit-modal_1rntr_47"))
                    )
                    print(f"ğŸ” iframe {i}ì—ì„œ ìš”ì†Œ ë°œê²¬!")
                    context_found = True
                    break
                except TimeoutException:
                    driver.switch_to.default_content()
            if not context_found:
                print("âš ï¸ ëª¨ë“  ì»¨í…ìŠ¤íŠ¸ì—ì„œ ìš”ì†Œë¥¼ ì°¾ì§€ ëª»í•¨")
                return

        # ë°ì´í„° ì…ë ¥
        for col_name, _, new_value in data:
            print(f"ğŸ” ì²˜ë¦¬ ì¤‘: {col_name}")
            try:
                
                elements = WebDriverWait(driver, 15).until(
                    EC.presence_of_all_elements_located((By.CLASS_NAME, "_columnBorder-edit-modal_1rntr_47"))
                )
                found = False
                for element in elements:
                    try:
                        time.sleep(2)
                        col_input = element.find_element(By.CSS_SELECTOR, 'input[type="text"][name$=".columnName"]')
                        current_col_name = col_input.get_attribute("value")
                        if current_col_name == col_name:
                            found = True
                            try:
                                # Ace Editor ì²˜ë¦¬ (JavaScript ë°©ì‹)
                                ace_editor = element.find_element(By.CSS_SELECTOR, "div.ace_editor")
                                driver.execute_script("arguments[0].click();", ace_editor)  # í¬ì»¤ìŠ¤ ì´ë™
                                time.sleep(0.5)  # í¬ì»¤ìŠ¤ í›„ ëŒ€ê¸°
                                set_ace_editor_value(driver, ace_editor, new_value)
                            except NoSuchElementException:
                                print(f"âš ï¸ {col_name} Ace Editorê°€ ì—†ìŒ, ì¼ë°˜ textarea ì²˜ë¦¬")
                                try:
                                    # textarea ì²˜ë¦¬ (í´ë¦½ë³´ë“œ ë°©ì‹)
                                    value_field = element.find_element(By.CSS_SELECTOR, 'textarea[name$=".dataValue"]')
                                    set_value_with_clipboard(driver, value_field, new_value)
                                except NoSuchElementException:
                                    print(f"âš ï¸ {col_name} ê°’ í•„ë“œê°€ ì—†ìŒ")
                            break
                    except StaleElementReferenceException:
                        print(f"âš ï¸ {col_name} StaleElement ë°œìƒ, ì¬íƒìƒ‰")
                        continue
                if not found:
                    print(f"âš ï¸ {col_name} ì°¾ì„ ìˆ˜ ì—†ìŒ")
            except TimeoutException:
                print(f"âš ï¸ {col_name} ìš”ì†Œ íƒìƒ‰ ì‹œê°„ ì´ˆê³¼")

        # ì €ì¥
        driver.switch_to.default_content()
        save_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//button[contains(text(), "ì €ì¥")]'))
        )
        save_button.click()
        print("âœ… ì €ì¥ ì™„ë£Œ!")
    except Exception as e:
        print(f"âš ï¸ ë°ì´í„° ë¶™ì—¬ë„£ê¸° ì‹¤íŒ¨: {e}")
        driver.switch_to.default_content()

# âœ… ë©”ì¸ ì‹¤í–‰
if __name__ == "__main__":
    excel_filename = "C:\\JJG\\Start\\sjdj\\scraped_data2.xlsx"
    print(f"ğŸ” í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬: {os.getcwd()}")
    if not os.path.exists(excel_filename):
        print(f"âš ï¸ íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤: {excel_filename}")
        exit()

    data = load_excel_data(excel_filename)
    if not data:
        print("âš ï¸ ë°ì´í„° ë¡œë“œ ì‹¤íŒ¨ë¡œ ì¸í•´ í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
        exit()

    driver1 = setup_browser(0)
    try:
        login_to_console(driver1)
        navigate_to_page(driver1)
        confirm_click(driver1)
        execute_tasks(driver1, 0, 5)
        paste_data(driver1, data)
    except KeyboardInterrupt:
        print("âš ï¸ ì‚¬ìš©ìê°€ í”„ë¡œê·¸ë¨ì„ ì¤‘ë‹¨í–ˆìŠµë‹ˆë‹¤.")
    finally:
        # driver1.quit()
        print("âœ… ë¸Œë¼ìš°ì €ê°€ ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
```
